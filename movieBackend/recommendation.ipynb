```python
movies=movies.merge(credits, on='title')
movies.head(4)
```

```python
movies=movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast
def token(text):
    listt=[]
    for i in ast.literal_eval(text):
        listt.append(i['name'])
    return listt
```

```python
movies['genres']=movies['genres'].apply(token)
```

```python
movies['keywords']=movies['keywords'].apply(token)
```

```python
movies.head()
```

```python
import ast
def cast_select(text):
    listt=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter<4:
            listt.append(i['name'])
        counter+=1
    return listt
```

```python
movies['cast']=movies['cast'].apply(cast_select)
```

```python
import ast
def crew_select(text):
    listt=[]
    counter=0
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            listt.append(i['name'])
            break
    return listt
```

```python
movies['crew']=movies['crew'].apply(crew_select)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].str.split()
```

```python
def no_space(word):
    listt=[]
    for i in word:
        new_word = i.replace(" ", "")
        listt.append(new_word)
    return listt
```

```python
movies['cast']=movies['cast'].apply(no_space)
movies['crew']=movies['crew'].apply(no_space)
movies['genres']=movies['genres'].apply(no_space)
movies['keywords']=movies['keywords'].apply(no_space)
```

```python
movies['tags']= movies['overview'] + movies['genres'] + movies ['keywords'] + movies['cast']
```

```python
movies.head()
```

```python
movies_tags=movies[['movie_id', 'title', 'tags']]
```

```python
movies_tags.head()
```

```python
movies_tags['tags']= movies_tags['tags'].apply(lambda x: " ".join(x))
```

```python
movies_tags['tags']= movies_tags['tags'].apply(lambda x: x.lower())
```

```python
import pandas as pd
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer(max_features=5000, stop_words="english")  # Limit features to 5000 for efficiency
```

```python
# Transform tags into vectorized form
movie_tfidf = vectorizer.fit_transform(movies_tags["tags"])  # Creates a sparse matrix
```

```python
# Convert sparse matrix to DataFrame
movie_tfidf = pd.DataFrame(movie_tfidf.toarray(), index=movies_tags["movie_id"])
```

```python
movie_tfidf.head()
```

```python
# Save the vectorized features to CSV
movie_tfidf.to_csv(r"C:\Users\HP\Documents\Projects\PookieCoders\backend\dataset\vectorized_movies.csv")
print("Vectorized data saved as CSV!")
```

```python
# Save the vectorized features as a Pickle file
with open(r"C:\Users\HP\Documents\Projects\PookieCoders\backend\dataset\vectorized_movies.pkl", "wb") as file:
    pickle.dump(movie_tfidf, file)
print("Vectorized data saved as Pickle!")
```
